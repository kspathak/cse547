```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_loss(ax, data, title):
    data_frame = pd.DataFrame(data)
    ax.plot(data_frame['step'], data_frame['training_loss'],
            label='training', c=plt.cm.Set1(0))
    ax.plot(data_frame['step'], data_frame['test_loss'],
            label='test', c=plt.cm.Set1(1))
    ax.plot(data_frame['step'], data_frame['validation_loss'],
            label='validation', c=plt.cm.Set1(2))
    ax.set_title(title)
    ax.set_xlabel('Steps')
    ax.set_ylabel('Mean-squared Error')
    ax.grid(True)
    ax.legend(title="Dataset")
    
def plot_misclassification_rate(ax, data, title):
    data_frame = pd.DataFrame(data)
    ax.plot(data_frame['step'], 1 - data_frame['training_accuracy'],
            label='training', c=plt.cm.Set1(0))
    ax.plot(data_frame['step'], 1 - data_frame['test_accuracy'],
            label='test', c=plt.cm.Set1(1))
    ax.plot(data_frame['step'], 1 - data_frame['validation_accuracy'],
            label='validation', c=plt.cm.Set1(2))
    ax.set_title(title)
    ax.set_xlabel('Steps')
    ax.set_ylabel('Misclassification Rate')
    ax.grid(True)
    ax.legend(title="Dataset")
```
